```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
N = 1950
t0 = 0
tmax = 0.015
dt = 0.001
def k(t):
    k = 0.000065
    return k
def p(t):
    p = 0.433
    return p
def dy(s,t):
    dy = ( k(t) + p(t)*s )*( N - s )
    return dy
t = np.arange(t0,tmax,dt) 
v0=25
s = odeint(dy,v0,t)
plt.plot(s[:,0],'r--', linewidth=2.0,label="решение")
plt.grid()
plt.show()
```
